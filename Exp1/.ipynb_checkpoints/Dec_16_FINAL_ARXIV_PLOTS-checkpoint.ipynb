# Part 1 --> plot predictions + CIs

```python
with open('TR_900/ffnn_yprs_50_runs.npy', 'rb') as f:
    ffnn_yprs50 = np.load(f)

with open('TR_900/naive_yprs_50_runs.npy', 'rb') as f:
    naive_yprs50 = np.load(f)

# plot mackie data, predictions and CI around these
data = pd.read_csv("../MackeyGlass_t17.dat", header=None).values.reshape(-1)

# Calculate CIs
used_curve = ffnn_yprs50[0]
lower = np.quantile(ffnn_yprs50 - used_curve, 0.025, axis=0)
upper = np.quantile(ffnn_yprs50 - used_curve, 0.975, axis=0)
used_curve2 = naive_yprs50[0]
lower2 = np.quantile(naive_yprs50 - used_curve2, 0.025, axis=0)
upper2 = np.quantile(naive_yprs50 - used_curve2, 0.975, axis=0)
```

```python
from matplotlib.pyplot import figure
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset


from IPython.display import display
from matplotlib import rc
#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)

rc('xtick', labelsize=20) 
rc('ytick', labelsize=20)
rc('font', size=20)#, usetex=True)
#rc('xlabel', labelsize=20) 
#rc('ylabel', labelsize=20)

# FFNN prediction curves + CI + mackie data
fig, ax = plt.subplots(figsize=(8,3), dpi=800)

#ax.setxlim()
ax.set_ylim(-0.6,0.6)
ax.set_xlim(1000,2350)
ax.plot(np.arange(1000,2000),data[1000:2000], 'k', alpha=0.8)
ax.plot(np.arange(1900,2000), used_curve, 'r',label="FFESN")
ax.plot(np.arange(1900,2000), used_curve - upper, 'm--') 
ax.plot(np.arange(1900,2000), used_curve - lower, 'm--')
ax.plot(np.arange(1900,2000),data[1900:2000], alpha=0.5)

# naive prediction curves + CI + mackie data
ax.plot(np.arange(1900,2000), used_curve2, 'b', label="ESN")
ax.plot(np.arange(1900,2000), used_curve2 - upper2, 'g--')
ax.plot(np.arange(1900,2000), used_curve2 - lower2, 'g--')
ax.set_xticks([1000, 1200, 1400, 1600, 1800, 2000])
ax.set_yticks([-0.4, -0.2, 0, 0.2, 0.4])
ax.set_xlabel(r'$t$')
ax.set_ylabel(r'$x(t)$')
axins = zoomed_inset_axes(ax, 4, loc='upper right') # zoom = 6
axins.set_ylim(0.05,0.175)
axins.set_xlim(1910,1950)
axins.plot(np.arange(1900,2000), used_curve, 'r')
axins.plot(np.arange(1900,2000), used_curve - upper, '--') 
axins.plot(np.arange(1900,2000), used_curve - lower, '--')
axins.plot(np.arange(1900,2000),data[1900:2000], 'k', alpha=0.8)
axins.plot(np.arange(1900,2000), used_curve2, 'b')
axins.plot(np.arange(1900,2000), used_curve2 - upper2, 'g--')
axins.plot(np.arange(1900,2000), used_curve2 - lower2, 'g--')
axins.set_xticks([1930])
axins.set_yticks([0.1])
rc('font', size=14)#, usetex=True)
#ax.set_xlabel(r'$t$')


#axins.imshow(Z2, extent=extent, interpolation="nearest",
#             origin="lower")
mark_inset(ax, axins, loc1=2, loc2=3)#, loc1=0.5, loc2=4, fc="none", ec="0.5")
rc('text', usetex=False)
ax.legend(loc=4, fontsize=15)
plt.draw()
plt.savefig("paper2plot1.pdf",bbox_inches='tight',transparent=True)
plt.show()
```

# Part 2 --> plot distribution of MSEs

```python
import numpy as np
import seaborn as sns
import scipy.stats
import pandas as pd
import matplotlib.pyplot as plt

with open('TR_900/ffnn_mses_50_runs.npy', 'rb') as f:
    ffnn_mses50 = np.load(f)
with open('TR_900/naive_mses_50_runs.npy', 'rb') as f:
    naive_mses50 = np.load(f)
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h

ffnn_CI = mean_confidence_interval(ffnn_mses50)
print(ffnn_CI)
lower_ffnnci = ffnn_CI[0]
upper_ffnnci = ffnn_CI[2]

naive_CI = mean_confidence_interval(naive_mses50)
print(naive_CI)
lower_naiveci = naive_CI[0]
upper_naiveci = naive_CI[2]

arrs = [ffnn_mses50, naive_mses50]
```

```python
#from matplotlib.pyplot import figure
import matplotlib as mpl
from matplotlib import rc
rc('xtick', labelsize=15) 
rc('ytick', labelsize=15)
rc('font', size=15)
rc('text', usetex=False)
rc('font', size=15,**{'family': 'Times New Roman', 'serif': ['Computer Modern']})
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1,figsize=(6,4),dpi=250,sharex=True)
#rc('text', usetex=True)
#ax2.set_xticklabels(["$%d$"%f for f in np.arange(0,11,2)])
#rc('text', usetex=False)

sns.histplot(ax=ax1,data=ffnn_mses50, kde=False, color="crimson", label="FFESN")
sns.histplot(ax=ax2,data=naive_mses50, kde=False, color="dodgerblue",label="ESN")
ax1.set_ylabel('Count')
ax2.set_xlabel('Mean-Squared Error')
ax1.legend(fontsize=13)
ax2.legend(fontsize=13)
#for label in ax1.get_yticklabels():
#    rc('text', usetex=True)
plt.savefig("paper2plot2.pdf",bbox_inches='tight',transparent=True)
plt.show()

import numpy as np
from scipy.stats import wilcoxon


# Perform Wilcoxon signed-rank test
stat, p = wilcoxon(ffnn_mses50, naive_mses50)

print(f"Statistic: {stat}")
print(f"P-value: {p}")

# Interpretation
alpha = 0.05
if p > alpha:
    print("There's no significant difference between the two samples.")
else:
    print("There's a significant difference between the two samples.")
```

# BOXPLOTS, TRAIN 900 AND 300

```python
from IPython.display import display
from matplotlib import rc
import numpy as np
import scipy.stats
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager
rc('text', usetex=False)
rc('xtick', labelsize=16) 
rc('ytick', labelsize=16)
rc('font', size=18)#, usetex=True)
rc('font', **{'family': 'Times New Roman', 'serif': ['Computer Modern']})

with open('TR_900/ffnn_mses_50_runs.npy', 'rb') as f:
    ffnn1_mses50 = np.load(f)

with open('TR_900/naive_mses_50_runs.npy', 'rb') as f:
    naive1_mses50 = np.load(f)

with open('TR_300/SMALL_TR_ffnn_mses_50_runs.npy', 'rb') as f:
    ffnn2_mses50 = np.load(f)

with open('TR_300/SMALL_TR_naive_mses_50_runs.npy', 'rb') as f:
    naive2_mses50 = np.load(f)

arrs1 = [ffnn1_mses50, naive1_mses50]
arrs2 = [ffnn2_mses50, naive2_mses50]

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2,figsize=(9,3),dpi=450,sharey=False)


my_colors = ["dimgrey", "lightgrey"]

box1 = sns.boxplot(ax=ax1, data=arrs1, palette=my_colors)
ax1.set_xticklabels(["FFESN","ESN"],size=18)
sns.boxplot(ax=ax2, data=arrs2, palette=my_colors)
ax2.set_xticklabels(["FFESN","ESN"],size=18)
ax2.yaxis.tick_right
ax1.set_ylabel("Mean-Squared Error",size=18)
ax1.set_title("900 Training Inputs")
ax2.set_title("300 Training Inputs")

plt.show()
```
